```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a NumPy array
data = np.random.rand(10, 2)

# Create a Pandas DataFrame
df = pd.DataFrame(data, columns=['x', 'y'])

# Create a scatter plot
plt.scatter(df['x'], df['y'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')
plt.show()
```
